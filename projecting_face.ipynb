## 놓친 부분

얼굴 인식을 시도하던 중 놓친 부분이 하나 있었다.

크롤링과 영상에서 추출한 이미지 데이터에서 얼굴을 크로핑했다.

그리고나서 크로핑한 얼굴의 눈코입이 가운데에 정렬되도록 'Projecting' 해야 하는데 이 부분을 놓친 것이다.
